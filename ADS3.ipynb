**HeatMap**

```python
data = pd.read_csv('diabetes.csv')
# Select features for visualization
features = ['Insulin', 'BMI']
data_subset = data[features]
# Calculate correlation matrix
correlation_matrix = data_subset.corr()
# Create a figure with two subplots
plt.figure(figsize=(8, 4))
# Subplot 1: Correlation Matrix Heatmap
plt.subplot(1, 2, 1)
sns.heatmap(correlation_matrix, annot=True, linewidths=0.5)
plt.title("HeatMap")
# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

**Bubble Chart**

```python
# Load the diabetes dataset
data = pd.read_csv('diabetes.csv')
# Selecting the features for visualization
x = data['Age']
y = data['Glucose']
bubble_size = data['Insulin']
# Create a Bubble Chart with colored bubbles based on SkinThickness
plt.figure(figsize=(8, 4))
plt.scatter(x, y, s=bubble_size, cmap='viridis', alpha=0.5)
plt.title('Bubble Chart')
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.colorbar(label='Insulin')
plt.grid(True)
plt.show()
```

**Bar Plot**

```python
col_names = ['Pregnancies	', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age','Outcome']
pima = pd.read_csv("/content/diabetes.csv", header=None, names=col_names)
pima = pima.iloc[1: , :]
df = pd.DataFrame(pima)
independent_variable = 'BloodPressure'
target_variable = 'Glucose'
# Creating a bar plot
plt.bar(df[independent_variable], df[target_variable], color='blue')
```

**Pie Chart**

```python
# Count the number of occurrences of each unique value in 'Outcome' column
outcome_counts = data['Outcome'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(outcome_counts, labels=outcome_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Outcomes')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

**Histogram**

```python
sns.histplot(data=data, x='Glucose', kde=True)
plt.title('Histogram of Glucose')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
```
